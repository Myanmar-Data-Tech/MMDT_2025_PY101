# Exception Handling


### Q2) Run the following programs and identify the potential errors.
Use exception handling approach to handle and print all the potential erros.


```python
#Q2.1)
a = 1
try:
  b = int(input("Please enter a number to divide a "))
  a = a/b
except ValueError:
  print("The denominator must be a number.")
else:
  print("Success a= ",a)
```

```python
#Q2.2)
mylist = [2, 3, 5, -7, 8, 0]
try:
  idx = 0
  b = 5
  while idx<=6:
    print('Iteration:', idx, mylist[idx]/b)
    idx = idx+1
except IndexError:
  print("Input is out of range.")
else:
  print('Iteration completed')




```
