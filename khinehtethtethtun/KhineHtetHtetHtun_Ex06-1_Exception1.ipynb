# Exception Handling


### Q1) Run the following programs and identify the errors.
Use exception handling approach to handle the erros.


```python
#Q1.1) Zero division error
def myfunction(a,b):
    try:
      answer = a/b
      return answer
    except ZeroDivisionError:
      print(f"Denominator value is {b}")


#Call the function as follows:
myfunction(2,0)
```

```python
#Q1.2 Name Error

def myfunction2(c):
  try:
    answer = a/c
    return answer
  except ZeroDivisionError:
    print("Denominator is zero")
  except NameError:
    print("Numerator is not defined.")


# Call the function as follows:
myfunction2(3)
```

```python
#Q1.3 IndexError
a = [1, 2, 3]
try:
  for k in range(0,4):
    print(a[k])
except IndexError:
  print("Index is out of range.")
```

```python
#Q1.4 Zero Division Error, Value Error
a = 1
try:
  b = int(input("Please enter a number to divide a"))
  a = a/b
except ZeroDivisionError:
  print("The denominator is zero")
except ValueError:
  print("Input is not a number")
else:
  print("Success a=",a)
```

```python
def divide(a, b):
    try:
        result = a / b
        print('The result of division is', result)
    except ZeroDivisionError:
        print('Numerator must be greater than zero')
    except TypeError:
        print('Inputs must be numbers')
    else:
        return result


divide("10",2)
```

```python
def read_file(filename):
    try:
        with open(filename, 'r') as file1:
            word_string = file1.read()
        print("File reading completed successfully")
        return word_string
    except FileNotFoundError:
        print("File does not exist.")



read_file("./data/Q1.txt")
```
