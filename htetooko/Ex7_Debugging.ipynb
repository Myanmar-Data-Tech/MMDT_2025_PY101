The following program should counts how many of the squares of the numbers from 1 to 100 end in 1.  (Example: 1,81,121). Debug the code.

```python
import numpy as np
def fun_square_count():
    num_list = np.arange(1,101)
    print("Original number is ", num_list)
    squred_num = [value**2 for value in num_list]
    print("Squared num is ", squred_num)
    count=0
    for num in squred_num:
        if(num//10) == 1:      
            count +=1
            
    return count

fun_square_count()
```

Debug the following programs

```python
def add(a, b):
    try:
        return a + int(b)
    except TypeError:
        return("Cannot add values of different type")

print(add(5, '2'))
```

```python
def divide(a, b):
    try:
        result = a / b
        print("result is ", result)
        return result
    except ZeroDivisionError:
        print("Cannot divided by zero")
    except TypeError:
        print("Both inputs must be int ot float")

divide(10,2)
divide(10,0)
divide(10,'a')
```

```python
def is_prime(n):
    if n <= 0:
        print(f"{n} is zero or minus")
        return False
    for i in range(2, n):
        if n % i == 0:
            print(f"{n} is divisible by {i} not a prime number")
            return False
    print(f"{n} is a prime number")
    return True

print(is_prime(15))
```
